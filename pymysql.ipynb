```python
pd.read_sql_query("""with cte1 as
(select gender,label, count(label) g_count
from credit_card cc join credit_card_label cl on cc.Ind_ID=cl.Ind_ID where gender  in ('M','F') 
group by GENDER,label), 
cte2 as
(select gender,sum(cte1.g_count)  as count_g  from cte1 group by gender)
select cte1.gender,cte2.gender,cte1.label,(cte1.g_count/cte2.count_g)*100 percentage from cte1,cte2
where cte1.gender=cte2.gender""",conn )
```
